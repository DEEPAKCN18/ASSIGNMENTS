#### Use Random Forest to prepare a model on fraud data treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
fc = pd.read_csv("Fraud_check.csv")
```

```python
fc
```

```python
fc=pd.get_dummies(fc,columns=['Undergrad','Marital.Status','Urban'], drop_first=True)
```

```python
fc["TaxInc"] = pd.cut(fc["Taxable.Income"], bins = [10002,30000,99620], labels = ["Risky", "Good"])
```

#### Lets assume: taxable_income <= 30000 as “Risky=0” and others are “Good=1”

```python
fc = pd.get_dummies(fc,columns = ["TaxInc"],drop_first=True)
```

```python
fc.head(10)
```

```python
fc.tail(10)
```

```python
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
fc_norm = norm_func(fc.iloc[:,1:])
```

```python
fc_norm.head(10)
```

```python
fc_norm.tail(10)
```

```python
X = fc_norm.drop(['TaxInc_Good'], axis=1)
y = fc_norm['TaxInc_Good']
```

```python
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print('Train accuracy: {}'.format((Xtrain, ytrain)))
print('Test accuracy: {}'.format((Xtest, ytest)))
```

```python
forest_new = RandomForestClassifier(n_estimators=100,max_depth=10,min_samples_split=20,criterion='gini')
forest_new.fit(Xtrain, ytrain)
```
