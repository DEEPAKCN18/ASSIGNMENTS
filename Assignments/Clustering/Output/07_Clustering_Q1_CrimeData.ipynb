```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
cd=pd.read_csv('crime_data.csv')
```

```python
cd.head()
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
cd_norm=norm_func(cd.iloc[:,1:])
```

```python
cd_norm
```

```python
dendrogram=sch.dendrogram(sch.linkage(cd_norm,method='average'))
```

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
cd3=pd.read_csv('crime_data.csv')
```

```python
cd3.info()
```

```python
cd3.drop(['Unnamed: 0'],axis=1,inplace=True)
cd3
```

```python
cd3_norm=StandardScaler().fit_transform(cd3)
cd3_norm
```

```python
dbscan=DBSCAN(eps=1,min_samples=4)
dbscan.fit(cd3_norm)
```

```python
dbscan.labels_
```

```python
cd3['clusters']=dbscan.labels_
cd3
```

```python
cd3.groupby('clusters').agg(['mean']).reset_index()
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
```

```python
cl
```

```python
pd.concat([cd3,cl],axis=1)
```
